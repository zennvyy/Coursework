```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
```

```python
def f1(x):
    return np.exp(-x**2) * (np.cos(2*x)**2 * np.sin(5*x) + np.sin(x)**2 + 1)
def f2(x):
    return 2 - np.arctan(x)**2
root = fsolve(f2, [-1, 1])
def rejection_sampling(f, a, b, iter=100000):
    samples = []

    for i in range(iter):
        x = np.random.uniform(a, b)
        y = np.random.uniform(0, 2)

        if y < f(x):
            samples.append(x)

    return np.array(samples)
```

```python
samples1 = rejection_sampling(f1, -2, 2)
samples2 = rejection_sampling(f2, root[0], root[1])
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(samples1, bins=30, density=True, alpha=0.6, color='g')
x = np.linspace(-2, 2, 100)
plt.plot(x, f1(x))
plt.title('f(x) = e^-x^2 (cos^2(2x)sin(5x)+sin^2(x)+1); {-2<x<2}')
plt.subplot(1, 2, 2)
plt.hist(samples2, bins=30, density=True, alpha=0.6, color='g')
x = np.linspace(root[0], root[1], 100)
plt.plot(x, f2(x))
plt.title('f(x) = 2 - arctan^2 (x); {axb} where (a,b) are roots of f(x)')
plt.show()
```
