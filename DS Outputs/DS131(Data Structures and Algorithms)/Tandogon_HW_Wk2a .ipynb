Graph the functions $8n$, $4n\space log\space n$, $2n^2$, $n^3$, and $2^n$ using a logarithmic scale for the $x-$ and $y-$axes; that is, if the function value $f (n)$ is $y$, plot this as a point with $x-$coordinate at $log \space n$ and $y-$coordinate at $log \space y$.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the range of n values
n_values = np.arange(1, 100)
```

```python
# Define the functions
f1 = 8 * n_values
f2 = 4 * n_values * np.log(n_values)
f3 = 2 * n_values**2
f4 = n_values**3
f5 = 2**n_values

# Calculate the logarithmic values for x and y axes
log_n_values = np.log(n_values)
log_f1 = np.log(f1)
log_f2 = np.log(f2)
log_f3 = np.log(f3)
log_f4 = np.log(f4)
log_f5 = np.log(f5)
```

```python
# Create subplots for each function
fig, axes = plt.subplots(5, 1, figsize=(10, 16))

# Plot each function on a separate graph
axes[0].plot(log_n_values, log_f1, label='8n', marker='o')
axes[0].set_title('Logarithmic Scale Plot of 8n')
axes[0].set_xlabel('log(n)')
axes[0].set_ylabel('log(8n)')

axes[1].plot(log_n_values, log_f2, label='4n log(n)', marker='o')
axes[1].set_title('Logarithmic Scale Plot of 4n log(n)')
axes[1].set_xlabel('log(n)')
axes[1].set_ylabel('log(4n log(n))')

axes[2].plot(log_n_values, log_f3, label='2n^2', marker='o')
axes[2].set_title('Logarithmic Scale Plot of 2n^2')
axes[2].set_xlabel('log(n)')
axes[2].set_ylabel('log(2n^2)')

axes[3].plot(log_n_values, log_f4, label='n^3', marker='o')
axes[3].set_title('Logarithmic Scale Plot of n^3')
axes[3].set_xlabel('log(n)')
axes[3].set_ylabel('log(n^3)')

axes[4].plot(log_n_values, log_f5, label='2^n', marker='o')
axes[4].set_title('Logarithmic Scale Plot of 2^n')
axes[4].set_xlabel('log(n)')
axes[4].set_ylabel('log(2^n)')

# Add legends and adjust layout
for ax in axes:
    ax.legend()
    ax.grid(True)

plt.tight_layout()
plt.show()
```
